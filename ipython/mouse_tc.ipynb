## Metadata Functions

```python
def clean_cis_data(cis_data,metadata,remove = True):
    cis_names = {}
    all_ds_ids = metadata.index.values
    cis_ds = [c.split('-')[-1] for c in cis_data.columns.values]
    for d in cis_ds:    
        correct_ds = sorted([x for x in all_ds_ids if d==x or d[0:7]==x])
        if correct_ds:
            correct_ds = correct_ds[-1]
            cis_names[d] = metadata.loc[correct_ds,'Tissue']+';'+str(metadata.loc[correct_ds,'Time'])+ '-%s' %correct_ds
        else:
            cis_names[d] = d
    all_names = set([cis_names[r] for r in cis_names])
    all_ids = []
    for nm in all_names:

        gd = sorted([x for x in cis_data.columns.values if x.split('-')[-1] in cis_names 
                     and cis_names[x.split('-')[-1]]==nm])[-1]
        all_ids.append(gd)
    cis_data = cis_data[all_ids]
    
    
    cis_data = cis_data.rename(columns = {c:cis_names[c.split('-')[-1]] for c in cis_data.columns.values})
    cis_data[pd.isnull(cis_data)] = 0
    return cis_data
```

## Combine Data

```python
def normalize_trans(trans_data):
    avg_trans = np.mean(trans_data.values,axis = 1)
    for c in trans_data.columns.values:
        offset = np.mean(avg_trans/trans_data[c].values)
        trans_data[c] = trans_data[c]*offset
    return trans_data

def combined_replicate_data_using(dataframe,metadata,columns = ['Tissue','Time'],func = np.mean):
    replicate_ids = {}
    for ds in metadata.index.values:
        name = ';'.join([str(metadata.loc[ds,x]) for x in columns])
        if name not in replicate_ids: replicate_ids[name] = []
        correct_ds = [x for x in dataframe.columns.values if ds==x or ds[0:7]==x[0:7]]
        if len(correct_ds)!=1: 
            #print ds,correct_ds
            continue
        replicate_ids[name].append(sorted(correct_ds)[-1])
    outd = pd.DataFrame(columns = replicate_ids,index = dataframe.index.values)
    for nm in replicate_ids:
        if func == 'replicated':
            vls = np.sum(dataframe[replicate_ids[nm]].values,axis = 1)
            outd[nm] = dataframe[replicate_ids[nm][0]].values.astype(bool)
            #outd[nm] = vls >= max(1,min(2,len(replicate_ids[nm])))
        else:
            outd[nm] = func(dataframe[replicate_ids[nm]].values,axis = 1)
    return outd

def combined_replicate_data(dataframe,metadata,columns = ['Tissue','Time'],func = np.mean):
    replicate_ids = {}
    for ds in metadata.index.values:
        name = ';'.join([str(metadata.loc[ds,x]) for x in columns])
        if name not in replicate_ids: replicate_ids[name] = []
        correct_ds = [x for x in dataframe.columns.values if ds==x or ds[0:7]==x[0:7]]
        if len(correct_ds)!=1: print ds,correct_ds
        replicate_ids[name].append(sorted(correct_ds)[-1])
    outd = pd.DataFrame(columns = replicate_ids,index = dataframe.index.values)
    for nm in replicate_ids:
        if func == 'replicated':
            vls = np.sum(dataframe[replicate_ids[nm]].values,axis = 1)
            outd[nm] = dataframe[replicate_ids[nm][0]].values.astype(bool)
            #outd[nm] = vls >= max(1,min(2,len(replicate_ids[nm])))
        else:
            outd[nm] = func(dataframe[replicate_ids[nm]].values,axis = 1)
    return outd

def get_ds_names(dataframe,metadata,columns = ['Tissue','Time']):
    replicate_ids = {}
    for ds in metadata.index.values:
        name = ';'.join([ds]+[str(metadata.loc[ds,x]) for x in columns])
        if name not in replicate_ids: replicate_ids[name] = []
        correct_ds = [x for x in dataframe.columns.values if ds==x or ds[0:7]==x[0:7]]
        if len(correct_ds)!=1: 
            print ds,correct_ds
        #    continue
        replicate_ids[name].append(sorted(correct_ds)[-1])
    return {replicate_ids[n][0]:n for n in replicate_ids}
```

```python
def compare_cis2trans(trans_data,cis_data):
    c2g = motifs.cluster2gene_mapping(motifs.CLUSTER_MAPPINGS_FILE)
    cors = {}
    good_trans_values = [tf for tf in trans_data.index.values if tf in cis_data.index]
    good_cell_values = [tf for tf in trans_data.columns.values if tf in cis_data.columns.values]
    for tf in trans_data.index.values:
        if tf not in cis_data.index: 
            cors[tf] = (0,np.nan)
            continue
        cors[tf] = scipy.stats.pearsonr(cis_data.loc[tf,good_cell_values].values,
                                        trans_data.loc[tf,good_cell_values].values)
    return cors
```

```python
#tissues = ['forebrain','mfBrain','midbrain','hindbrain','mBrain','mCerebrum','mCerebellum'] #,'retina','mRetina']
#fdr_peaks = rep_peak_data.index.values[fdrs<2]

time_conversion = {'nan':30,'P0':21,'P7':28,'P12':31}
#metadata['Time'][pd.isnull(metadata['Time'])] = 30
#metadata['Time'][metadata['Time'].values=='P0'] = 21
#metadata['Time'][metadata['Time'].values=='P7'] = 28
#metadata['Time'][metadata['Time'].values=='P12'] = 31
#metadata['Time'][metadata['Time'].values=='e14.5d'] = 14.5
#metadata['Time'][metadata['Time'].values=='e10.5'] = 10.5
#metadata['Time'] = metadata['Time'].values.astype(float)


def tissue_dhs_order(tissues,out_base,fdr_peaks = None,tissue_preset = defaultdict(int),rep_min = 2,fold_change =2):
    tissue_order = sorted([x for x in rep_peak_data.columns.values if x.split(';')[0] in tissues],
                          key = lambda k: tissue_preset[k.split(';')[0]])
    
    time_values = []
    for x in tissue_order:
        try: 
            time_values.append(float(x.split(';')[1]))
        except ValueError:
            time_values.append(time_conversion[x.split(';')[1]])
        if np.isnan(time_values[-1]): time_values[-1] = 30
    tissue_order = np.array(tissue_order)[np.argsort(time_values,kind = 'mergesort')]
    time_vector = np.sort(time_values)
    tissue_vectors = np.array([tissues.index(r.split(';')[0])*.01 for r in tissue_order])
    
    all_ids = [x for x in metadata.index.values if any([metadata.loc[x,'Tissue']==t for t in tissues])] 
    call_peaks = ((np.sum(call_data[all_ids].values,axis = 1)>=rep_min) & 
                (np.sum(call_data[all_ids].values,axis = 1)<=len(all_ids)))
    call_peaks = call_peaks & ((np.max(np.log(pk_data[all_ids].values+1.0),axis=1)-
                                np.mean(np.log(pk_data[all_ids].values+1.0),axis=1))>np.log(fold_change))
    
    print '%dCall Peaks' %np.sum(call_peaks)
    print len(all_ids),'%s' %tissues[0]
    if fdr_peaks is None: fdr_peaks = np.ones(len(call_peaks)).astype(bool)
    good_peaks = call_data.index.values[call_peaks & fdr_peaks]
    weights = rep_peak_data.loc[good_peaks,tissue_order].values
    wavg = np.sum(weights*time_vector,axis = 1) / np.sum(weights,axis = 1)


    worder = np.argsort(wavg)
    max_point = np.argmax(rep_peak_data.loc[good_peaks,tissue_order].values[worder],axis = 1)
    for tp in set(list(max_point)):
        name = tissue_order[tp]
        write_subset(rep_peak_data.index.values[good_peaks][max_point==tp],
                     MASTER_LIST,out_base + '.%s.peaks.bed' %name)
    
    print time_vector,tissue_order,tissue_vectors
    pk_scores = time_vector[max_point]+tissue_vectors[max_point]
    morder = np.argsort(pk_scores,kind = 'mergesort')
    order = good_peaks[worder][morder]
    pk_scores = time_vector[max_point]
    return order,tissue_order,pk_scores

def tissue_heatmap(tissues,out_base,fdr_peaks = None, figsize = (5,6),xoffset = -0.5,set_rot = 'vertical',
                   vmin = -.5,vmax = 2,tissue_preset = defaultdict(int),top_color = sns.xkcd_rgb['blue']):

    dhs_order,tissue_order,pk_scores = tissue_dhs_order(tissues,out_base,fdr_peaks = fdr_peaks,tissue_preset = tissue_preset)



    zscores = rep_peak_data.loc[dhs_order,tissue_order]
    std_values = np.std(zscores.values,axis = 1)
    mean_values = np.mean(zscores.values,axis = 1)
    zscores = np.log(zscores+1)
    zscores = zscores.subtract(np.mean(zscores.values,axis = 1),axis = 'rows')
    #zscores = zscores.subtract(mean_values,axis = 'rows')
    #zscores = zscores.divide(std_values,axis = 'rows')


    #change = {c:(c[0],'P0') for c in zscores.columns.values if c[1]==21}
    #for c in zscores.columns.values:
    #    if c[1]>22 and c[1]!='P0': change[c] = (c[0],'Adult')
    #zscores = zscores.rename(columns = change)
    zscores= zscores.rename(columns = {c:c.replace(';','-') for c in zscores.columns.values})
    zscores= zscores.rename(columns = {c:c.replace('nan','Adult') for c in zscores.columns.values})

    altius = sns.color_palette(["#333333", "#F1CC81"])
    from matplotlib.colors import ListedColormap
    my_cmap = ListedColormap(altius.as_hex())
    my_cmap = sns.blend_palette(["#333333","#F1CC81"],n_colors = 20,as_cmap = True)
    #my_cmap = 'Blues'

    with sns.axes_style('white'):
        f1 = plt.figure(figsize = figsize)
        ax = plt.subplot(111)
        print '%d DHS Order' %len(dhs_order)
        r = sns.heatmap(zscores.loc[dhs_order],cmap = my_cmap,linewidths = 0,rasterized = True,vmin = vmin,
                    vmax = vmax,yticklabels = False,xticklabels = True)

        #ax.set_ylabel('%s DHSs' %len(good_peaks),color = 'orange')
        ax.set_ylabel('')
        ax.set_xlabel('')
    
        start = 0
        top,bottom = ax.get_ylim()
        ax.set_xlim([xoffset,ax.get_xlim()[1]])
        for vls in sorted(set(list(pk_scores))):
            nmb = np.sum(pk_scores==vls)
            #print vls,nmb
            length = nmb
            #print top,length,start
            ax.plot([-.1,-.1],[start+0.1*length,start+.9*length],lw = 3,color = top_color)
            ax.text(-.15,start + .5*length,'%d DHSs' %nmb,rotation = set_rot,color = top_color,
                    verticalalignment='center',horizontalalignment = 'right',fontsize = 8)
            start += nmb
        ax.text(xoffset,.5*top,'%s DHSs' %(len(dhs_order)),rotation = 'vertical',color = top_color,
                    verticalalignment='center',horizontalalignment = 'right',fontsize = 12)
        plt.subplots_adjust(left = 0.065,bottom = 0.22)
    return f1

MASTER_LIST='master-peaks.mouse.bed4'
def write_subset(subset,bed_file,output_file):
    count = 0
    subset = set([x for x in subset])
    fout = open(output_file,'w')
    with open(bed_file) as f1:
        for line in f1:
            if count in subset: fout.write(line)
            count += 1
    fout.close()
```

```python
def compare_cis2trans(trans_data,cis_data):
    c2g = motifs.cluster2gene_mapping(motifs.CLUSTER_MAPPINGS_FILE)
    cors = {}
    good_trans_values = [tf for tf in trans_data.index.values if tf in cis_data.index]
    good_cell_values = [tf for tf in trans_data.columns.values if tf in cis_data.columns.values]
    for tf in trans_data.index.values:
        if tf not in cis_data.index: 
            cors[tf] = (0,np.nan)
            continue
        cors[tf] = scipy.stats.pearsonr(cis_data.loc[tf,good_cell_values].values,
                                        trans_data.loc[tf,good_cell_values].values)
    return cors

def normalize_trans(trans_data):
    avg_trans = np.mean(trans_data.values,axis = 1)
    for c in trans_data.columns.values:
        offset = np.mean(avg_trans/trans_data[c].values)
        trans_data[c] = trans_data[c]*offset
    return trans_data



def combined_replicate_data(dataframe,metadata,columns = ['Tissue','Time']):
    replicate_ids = {}
    for ds in metadata.index.values:
        name = ';'.join([str(metadata.loc[ds,x]) for x in columns])
        if name not in replicate_ids: replicate_ids[name] = []
        correct_ds = [x for x in dataframe.columns.values if ds==x or ds[0:7]==x[0:7]]
        #if len(correct_ds)!=1: print ds,sorted(correct_ds)
        replicate_ids[name].append(sorted(correct_ds)[-1])
    replicates = replicate_ids.keys() #[x for x in replicate_ids if len(replicate_ids[x])>1]
    outd = pd.DataFrame(columns = replicates,index = dataframe.index.values)
    for nm in replicates:
        outd[nm] = np.mean(dataframe[replicate_ids[nm]].values,axis = 1)
    return outd
```

```python
def tissue_tf_heatmap(tissue_name,trans_zscores,cis_zscores,zscore_min = 2,number = 10,combined_zscore = 3):
    
    cvalues = [x for x in trans_zscores.columns.values if any([r in x for r in tissue_name]) 
               and 'glia' not in x.lower()]
    
    trans_sign = np.ones(trans_zscores.shape[0])
    for i,tf in enumerate(trans_zscores.index.values): 
        if tf not in tf_order: continue
        if tf_fdrs[tf_order.index(tf)]<0.01 and tf_cors[tf][0]<0: trans_sign[i]=-1
        if tf_fdrs[tf_order.index(tf)]<0.01 and tf_cors[tf][0]>0: trans_sign[i]=1
    combined_zscores = (trans_zscores[cvalues].values*trans_sign[:,np.newaxis] + 
                        cis_zscores.loc[trans_zscores.index.values,cvalues].values)
    combined_zscores /= np.sqrt(2.)
    combined_zscores[np.isnan(combined_zscores)] = 0
    combined_zscores = pd.DataFrame(combined_zscores,
                                index = trans_zscores.index.values,columns = [cvalues]) 
    
    
    signed_trans = trans_zscores[cvalues].multiply(trans_sign,axis = 'rows')
    tissue_tfs = get_tfs(combined_zscores,signed_trans,cis_zscores[cvalues],
                         zscore_min = zscore_min,number = number,combined_min = combined_zscore)
    
    f1 = tf_heatmap(tissue_tfs,trans_data_tf,cis_data,combined_zscores)
    return f1
    
def get_tfs(combined_zscores,trans_zscores,cis_zscores,zscore_min = 2,combined_min = 2,number = 10):
    
    maxv = np.max(combined_zscores.values,axis = 1)
    stdv = np.std(combined_zscores.values,axis = 1)
    
    max_cis = np.max(cis_zscores.loc[combined_zscores.index.values].values,axis = 1)
    max_trans = np.max(trans_zscores.loc[combined_zscores.index.values].values,axis = 1)
    
    good = (max_cis > zscore_min) & (max_trans > zscore_min)
    good = good & (maxv > combined_min)
    return combined_zscores.index.values[good][np.argsort(stdv[good])][::-1][0:number]


time_conversion = {'P0':21,'nan':30,'P7':28}
for x in ['10.5','11.5','14.5','18.5','11','9','10','9.5']: time_conversion[x] = float(x)

def tf_heatmap(developmental_tfs,tf_trans,tf_cis,combined):
    from matplotlib.colors import ListedColormap
    my_cmap = sns.blend_palette(["#333333","#F1CC81"],n_colors = 20,as_cmap = True)

    
    
    combined = combined.rename(columns = {r:'-'.join(r.split(';')) for r in combined.columns.values})
    tf_trans = tf_trans.rename(columns = {r:'-'.join(r.split(';')) for r in tf_trans.columns.values})
    tf_cis = tf_cis.rename(columns = {r:'-'.join(r.split(';')) for r in tf_cis.columns.values})
    

    tissue_order = sorted(combined.columns.values,key = lambda k: time_conversion[k.split('-')[1]])
    

    time_vector = sorted([r.split('-')[1] for r in tissue_order])
    time_vector = np.array([time_vector.index(r.split('-')[1]) for r in tissue_order])
    

    weights = combined.loc[developmental_tfs,tissue_order].values
    wavg = np.sum(weights*time_vector,axis = 1) / np.sum(weights,axis = 1)
    worder = np.argsort(wavg)
    
    max_point = np.argmax(combined.loc[developmental_tfs,tissue_order].values[worder],axis = 1)
    pk_scores = time_vector[max_point]#+tissue_vectors[max_point]
    morder = np.argsort(pk_scores,kind = 'mergesort')
    tf_order = developmental_tfs[worder][morder]


    f1 = plt.figure(figsize = (3.5,8))
    for i,tf in enumerate(tf_order):
        ax = plt.subplot(len(tf_order),1,i+1)
        
        rna_values = np.zeros(len(tissue_order)) + np.nan
        if tf in tf_rna.index.values:
            for xx,ts in enumerate(tissue_order):
                if ts.replace('-',';') in tf_rna.columns.values: rna_values[xx] = tf_rna.loc[tf,ts.replace('-',';')]
        if np.sum(np.isnan(rna_values))==len(rna_values): rna_values = None
        hd = np.vstack([tf_trans.loc[tf,tissue_order].values,tf_cis.loc[tf,tissue_order].values]).astype(float)
        hd[0] = (hd[0]-np.min(tf_trans.loc[tf].values))/(np.max(tf_trans.loc[tf].values)-np.min(tf_trans.loc[tf].values))
        hd[1] = (hd[1]-np.min(tf_cis.loc[tf].values))/(np.max(tf_cis.loc[tf].values)-np.min(tf_cis.loc[tf].values))
        names = ['Motif','TSS']
        if rna_values is not None:
            hd = np.vstack([hd,rna_values])
            hd[2] = hd[2]/float(np.max(tf_rna.loc[tf].values[~pd.isnull(tf_rna.loc[tf].values)]))#-np.min(tf_rna.loc[tf].values))
            names = ['Motif','TSS','RNA']
            
        #hd[1] = np.log2(hd[1]/np.mean(tf_cis.loc[tf].values))
        #hd = (hd-np.mean(hd,axis = 1)[:,np.newaxis])/np.std(hd,axis = 1)[:,np.newaxis]
        if i == len(tf_order)-1:
            sns.heatmap(hd,ax = ax,cbar = False,yticklabels = names,xticklabels= tissue_order,
                        cmap = my_cmap,linewidth = .5,vmax = .9,vmin = .1)            
        else:
            sns.heatmap(hd,ax = ax,cbar = False,yticklabels = names,xticklabels= False,
                        cmap = my_cmap,linewidth = .5,vmax = .9,vmin = .1)
        ax.yaxis.tick_right()
        ax.set_ylabel(tf)
        #yloc = plt.MaxNLocator(4)
        #ax.yaxis.set_major_locator(yloc)
    plt.tight_layout()
    return f1


if not os.path.exists('tc_tfs'): os.mkdir('tc_tfs')

f1 = tissue_tf_heatmap(['retina','mRetina'],trans_zscore,cis_zscore,zscore_min = 1,number = 10,combined_zscore = 2.5)
f1.savefig('tc_tfs/retina-tfs.pdf')
f1 = tissue_tf_heatmap(['forebrain'],trans_zscore,cis_zscore,zscore_min = 1,number = 10,combined_zscore = 2)
f1.savefig('tc_tfs/forebrain-tfs.pdf')
f1 = tissue_tf_heatmap(['mLung','lung'],trans_zscore,cis_zscore,zscore_min = 1,combined_zscore = 2)
f1.savefig('tc_tfs/lung-tfs.pdf')
f1 = tissue_tf_heatmap(['hindbrain'],trans_zscore,cis_zscore,zscore_min = 1,number = 10,combined_zscore = 2)
f1.savefig('tc_tfs/hindbrain-tfs.pdf')


f1 = tissue_tf_heatmap(['embryo'],trans_zscore,cis_zscore,zscore_min = 0,number = 10,combined_zscore = 2)
f1.savefig('tc_tfs/embryo-tfs.pdf')
f1 = tissue_tf_heatmap(['mHeart','heart'],trans_zscore,cis_zscore,zscore_min = 1,number = 10,combined_zscore = 2)
f1.savefig('tc_tfs/heart-tfs.pdf')
f1 = tissue_tf_heatmap(['craniofacial'],trans_zscore,cis_zscore,zscore_min = 1,combined_zscore = 2)
f1.savefig('tc_tfs/craniofacial-tfs.pdf')
```

```python
def gini_coefficient(values):
    if len(values)<3: return 0
    
    #sv = sorted(values)
    #return (sv[-1]-sv[-2])/np.sum(sv)
    total = 0
    if np.sum(values) == 0: return np.nan
    for i,v1 in enumerate(values):
        for j,v2 in enumerate(values):
            total += abs(v1-v2)
    return total / float(2*len(values)*np.sum(values))

def gini_coefficient2(values):
    if len(values)<3: return 0
    
    sv = np.sort(values)
    top = 2*np.sum((np.arange(len(sv))+1)*sv)
    bottom = float(len(sv)*np.sum(sv))
    
    return top/bottom - (len(sv)+1)/float(len(sv))


def get_motif_gini(cis_data,motif_data,zscore_limit = 3):
    m2g = motifs.cluster2gene_mapping(motifs.CLUSTER_MAPPINGS_FILE)
    cells = motif_data.columns.values
    names = []
    cvs = []
    ginis = []
    for m in m2g:
        mn = int(m)
        if mn not in motif_data.index: 
            continue
        mean = np.mean(motif_data.loc[mn].values)
        std = np.std(motif_data.loc[mn].values)
        zscores = (motif_data.loc[mn].values - mean)/std
        
        if len(m2g[m]) < 3:
            continue
        if max(zscores)<zscore_limit: 
            continue
        
        names.append(m)
        cvs.append(np.std(motif_data.loc[mn])/np.mean(motif_data.loc[mn]))
        these = []
        for i,zs in enumerate(zscores):
            if zs <= zscore_limit: continue
            these.append(gini_coefficient(cis_data.loc[[x for x in m2g[m] if x in cis_data.index],cells[i]]))
        ginis.append(np.mean(these))
            
            
    return cvs,ginis,names

def skewness(values):
    std = np.std(values)
    mn = np.mean(values)
    offset = values-mn
    return np.mean((offset/std)**3)

def kurtosis(values,directed = False):
    std = np.std(values)
    mn = np.mean(values)
    offset = values-mn
    kurt = np.mean((offset/std)**4)
    #if directed:
    #    kurt = np.mean(offset[offset>np.median(offset)]**4)/std**4
    #else:
    #    kurt = np.mean(offset**4)/std**4
    return kurt
```

```python
def get_ds_names(dataframe,metadata,columns = ['Tissue','Time']):
    replicate_ids = {}
    for ds in metadata.index.values:
        name = ';'.join([ds]+[str(metadata.loc[ds,x]) for x in columns])
        if name not in replicate_ids: replicate_ids[name] = []
        correct_ds = [x for x in dataframe.columns.values if ds==x or ds[0:7]==x[0:7]]
        #if len(correct_ds)!=1: 
            #print ds,correct_ds
        #    continue
        replicate_ids[name].append(sorted(correct_ds)[-1])
    return {replicate_ids[n][0]:n for n in replicate_ids}
```

```python
def match_metadata(cis_data,metadata,remove = True):
    cis_names = {}
    all_ds_ids = metadata.index.values
    cis_ds = cis_data.columns.values
    for d in all_ds_ids:    
        correct_ds = sorted([x for x in cis_ds if d==('DS'+x.split('DS')[-1])
                             or d[0:7]==('DS' + x.split('DS')[-1][0:5])])
        if correct_ds:
            correct_ds = correct_ds[-1]
            cis_names[correct_ds] = metadata.loc[d,'Tissue']+';'+str(metadata.loc[d,'Time'])+ '-%s' %d
        else:
            print d,correct_ds
    cis_data = cis_data[cis_names.keys()]
    cis_data = cis_data.rename(columns = {c:cis_names[c] for c in cis_data.columns.values})
    cis_data[pd.isnull(cis_data)] = 0
    return cis_data
```

```python
def landscape_plot(plot_values,order = None,ax = None):
    normed = -1*plot_values
    if order is None:
        order = middle_out(normed)
    
    if ax is None:
        f1 = plt.figure()
        ax = plt.subplot(111)
        
    for i,x in enumerate(order):
        ax.plot([i,i],[0,normed[x]],color = 'grey',linewidth = 0.5)
        #ax.plot([0,-1*normed[x]],[i,i],color = 'grey',linewidth = 0.5)
    ax.plot(np.arange(len(order)),normed[order.astype(int)],color = 'k',linestyle = '',marker = 'o',markersize = 3,
            markeredgewidth = 0)
    #ax.plot(-1*normed[order.astype(int)],np.arange(len(order)),color = 'k',linestyle = '',marker = 'o',markersize = 3,
    #        markeredgewidth = 0)
    return order
def middle_out(normed):
    order = np.zeros(len(normed))
    s_order = np.argsort(normed)
    mid = len(order)/2
    for i,ind in enumerate(s_order):
        offset = (i)/2 * (i/2==i/2.) - (i/2+1) * (i/2 != i/2.)
        order[mid + offset] = ind 
    return order
```

```python
time_conversion = {'10.5':0,'11.5':1,'14.5':2,'P0':3}
def create_regress_data(tf,dataframe,metadata):
    out_data = {'Time':[],'Tissue':[],'Score':[]}
    count = 0
    for x in dataframe.columns.values:
        #if x not in metadata.index.values: continue
        #name = metadata.loc[x,'Tissue']
        #time = metadata.loc[x,'Time']
        #print x
        name,time = x.split('-')[0].split(';')
        if 'brain' not in name: continue
        out_data['Tissue'].append(name)
        out_data['Time'].append(time)
        out_data['Score'].append(dataframe.loc[tf,x])
        count += 1
    return pd.DataFrame.from_dict(out_data)

def plot_regress_data(regress_data,ax = None,vmax = 3):
    tissues = ['forebrain','midbrain','hindbrain']
    times = ['10.5','11.5','14.5','P0']
    pdata = pd.DataFrame(np.zeros((3,4)),index = tissues,columns = times)
    for ts in tissues:
        for tm in times:
            pdata.loc[ts,tm] = np.mean(regress_data['Score'].values[
                                        (regress_data['Tissue'].values==ts) & (regress_data['Time'].values==tm)])    
    sns.heatmap(pdata,cmap = 'RdBu_r',vmax = vmax,vmin = -vmax,ax = ax,cbar = False)


import statsmodels.formula.api as sm
from statsmodels.stats.api import anova_lm

def run_regression(regress_data):
    regress_data['cTime'] = [time_conversion[t] for t in regress_data['Time'].values]
    f1 = sm.ols(formula="Score ~ Tissue + cTime + Tissue:cTime", data=regress_data).fit()
    f2 = sm.ols(formula="Score ~ Tissue", data=regress_data).fit()
    f3 = sm.ols(formula="Score ~ cTime", data=regress_data).fit()
    f4 = sm.ols(formula="Score ~ 1", data=regress_data).fit()

    time_pvalue = f1.compare_f_test(f2)[1]
    tissue_pvalue = f1.compare_f_test(f3)[1]
    total_pvalue = f1.compare_f_test(f4)[1]
    #print f1.summary()
    return tissue_pvalue,time_pvalue
```

```python
def get_optimal_mean_filter(means,pvalues,threshold):
    from statsmodels.sandbox.stats.multicomp import fdrcorrection0   
    if type(means) is not np.ndarray:
        means = np.array(means)
    if type(pvalues) is not np.ndarray:
        pvalues = np.array(pvalues)
    pvalues[np.where(np.isnan(pvalues))[0]] = 1.0  #### this is just to prevent the 
    #### final pvalues from being nan.  Shouldn't affect # passing threshold
    sort_index = np.argsort(means)
    current_number = 0#int(0.15*len(sort_index))
    count = 0
    max_number = 0
    passed,new_pvalues = fdrcorrection0(pvalues[sort_index[current_number:]],threshold)
    last_pass = np.sum(passed)
    direction = 1
    step_size = max(int(0.1*len(sort_index)),1)
    while step_size > 1 and count < 10000:
        count += 1
        proposal = max(current_number + step_size,0)
        proposal = min(proposal,len(pvalues)-1)
        test_pass,test_pvalues = fdrcorrection0(pvalues[sort_index[proposal:]],threshold)
        tpass = np.sum(test_pass)
        if tpass < last_pass:
            current_number = max(0,max_number - step_size)
            step_size = max(step_size/5,1)
            last_pass = 0
            max_number = current_number
        elif tpass > last_pass:
            max_number = proposal
            current_number = proposal
            last_pass = tpass
            passed = test_pass
            new_pvalues = test_pvalues
        else:
            current_number = proposal
    
    filter_mean = means[sort_index[current_number]]
    filter_index = np.searchsorted(means[sort_index],filter_mean)
    pass_peaks,new_pvalues = fdrcorrection0(pvalues[sort_index[filter_index:]],threshold)
    final_pvalues = np.ones(len(sort_index))
    final_pvalues[sort_index[filter_index:]] = new_pvalues
    return final_pvalues   
    ### this is going to take awhile   
```

```python
ensg2gene = pd.read_table('/home/jlazar/annotations/m9.ensg2gene.txt',header = None,index_col = 0,names = ['Gene'])
ensg2gene = ensg2gene.rename(index = {c:c.split('.')[0] for c in ensg2gene.index.values})

def average_rna_data(dt):
    individual = set([x.split('_day')[0] for x in dt.columns.values
                      if 'transcript_id' not in x])
    outd = pd.DataFrame(index = dt.index.values,columns = list(individual))
    for i in individual:
        outd[i] = np.mean(dt[[c for c in dt.columns.values if c.split('_day')[0]==i]].values,axis = 1)
    return outd
    
def drop_rna_dup(rna_data):
    max_values = {}
    rna_data = rna_data[~pd.isnull(rna_data.index)]
    dup_genes = list(set([x for x in rna_data.index.values[rna_data.index.duplicated()]]))
    for gene in dup_genes:
        best = np.argsort(-1 * np.max(rna_data.loc[gene].values[:,1:],axis = 1))[0]
        max_values[gene] = rna_data.loc[gene].values[best]
    rna_data = rna_data.loc[~rna_data.index.duplicated()]
    
    all_data = np.vstack([max_values[g] for g in dup_genes])
    for i,c in enumerate(rna_data.columns.values):
        rna_data.loc[dup_genes,c] = all_data[:,i]

    return rna_data
    
rna_data = pd.read_table('/home/cbreeze/forLazar/RNAseq_data/mouse_RNA_seq_data_matrix.fpkms',index_col = 0,header = 0)
rna_data = average_rna_data(rna_data)
rna_data = rna_data.rename(index = {c:c.split('.')[0] for c in rna_data.index.values if type(c) is str})
rna_data = rna_data.rename(index = {c:ensg2gene.loc[c,'Gene'].upper() for c in ensg2gene.index.values})

#rna_data = rna_data.subtract(np.mean(rna_data,axis = 1),axis = 'rows')
#rna_data = rna_data.divide(np.std(rna_data,axis = 1),axis = 'rows')
rna_data[pd.isnull(rna_data)] = 0
rna_data = drop_rna_dup(rna_data)
```
